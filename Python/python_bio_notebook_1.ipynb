## Section 6 — Counting with loops

A very common pattern:
- start a counter at 0
- loop through values
- increase the counter when a condition is true

```python
seq = "ATGCGCGT"

g_count = 0
for base in seq:
    if base == "G":
        g_count += 1

print("Number of G bases:", g_count)
```

## Section 16 — Functions: GC content

Here is a simple function for GC content.

```python
def gc_content(seq):
    gc = 0
    for base in seq:
        if base in "GC":
            gc += 1
    return gc / len(seq)

print(gc_content("ATGC"))
```

## Section 20 — Mini problem: sequence summary

This combines strings, counting, and f-strings.

```python
seq = "ATGCGTAA"

length = len(seq)
gc = sum(base in "GC" for base in seq) / len(seq)

print(f"Sequence: {seq}")
print(f"Length: {length}")
print(f"GC content: {gc:.2f}")
```

## Section 21 — Mini problem: multiple sequences

Loop through multiple sequences and summarize them.

```python
sequences = ["ATGC", "GGGG", "TTTT"]

for seq in sequences:
    gc = sum(base in "GC" for base in seq) / len(seq)
    print(f"{seq} | length={len(seq)} | gc={gc:.2f}")
```

## Section 23 — Error handling preview

Error handling helps code fail safely.
We will use it much more in Notebook 2.

```python
def safe_gc_content(seq):
    try:
        return sum(base in "GC" for base in seq) / len(seq)
    except ZeroDivisionError:
        return 0

print(safe_gc_content("ATGC"))
print(safe_gc_content(""))
```

## Section 25 — Final practice

Final mixed exercise.

```python
def summarize_sequence(seq):
    length = len(seq)
    gc = safe_gc_content(seq)
    counts = Counter(seq)
    print(f"Sequence: {seq}")
    print(f"Length: {length}")
    print(f"GC content: {gc:.2f}")
    print(f"Counts: {counts}")

summarize_sequence("ATGCCGTA")
```
