## Section 1 — Why Classes?
Instead of repeatedly calculating properties of sequences manually, we can represent sequences as objects.

```python
seq = "ATGCGT"

length = len(seq)
gc = sum(b in "GC" for b in seq) / len(seq)

print(length, gc)
```

## Section 2 — Your First Class

```python
class Sequence:

    def __init__(self, seq):
        self.seq = seq

    def length(self):
        return len(self.seq)

    def gc_content(self):
        return sum(b in "GC" for b in self.seq) / len(self.seq)


s = Sequence("ATGCGT")
print(s.length())
print(s.gc_content())
```

## Section 3 — Adding Methods

```python
class Sequence:

    def __init__(self, seq):
        self.seq = seq

    def reverse(self):
        return self.seq[::-1]

    def reverse_complement(self):
        comp = {"A":"T","T":"A","G":"C","C":"G"}
        return "".join(comp[b] for b in reversed(self.seq))


s = Sequence("ATGCGT")
print(s.reverse())
print(s.reverse_complement())
```

## Section 5 — Parsing FASTA into Objects

```python
def read_fasta(path):

    sequences = []

    with open(path) as f:

        name = None
        seq = []

        for line in f:
            line = line.strip()

            if line.startswith(">"):
                if name:
                    sequences.append(Sequence("".join(seq)))

                name = line[1:]
                seq = []

            else:
                seq.append(line)

        sequences.append(Sequence("".join(seq)))

    return sequences
```

## Section 7 — Variant Class

```python
class Variant:

    def __init__(self, gene, position, mutation):
        self.gene = gene
        self.position = position
        self.mutation = mutation


v = Variant("TP53", 273, "R273H")

print(v.gene)
print(v.position)
print(v.mutation)
```
