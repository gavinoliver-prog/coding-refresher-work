## Section 1 — Import Libraries
We will use pandas, matplotlib, and scikit-learn.

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

## Section 2 — Create a Toy Biological Dataset
This dataset mimics simple per-sample biological features and a responder label.

```python
data = {
    "sample": ["S1","S2","S3","S4","S5","S6","S7","S8"],
    "gc_content": [0.40, 0.55, 0.62, 0.35, 0.70, 0.48, 0.66, 0.30],
    "length": [100, 120, 150, 90, 180, 110, 160, 80],
    "motif_count": [1, 3, 4, 0, 5, 2, 4, 0],
    "responder": [0, 1, 1, 0, 1, 0, 1, 0]
}

df = pd.DataFrame(data)
df
```

## Section 3 — Define Features and Labels
In machine learning, features are your input columns and the label is what you want to predict.

```python
X = df[["gc_content", "length", "motif_count"]]
y = df["responder"]

print(X)
print(y)
```

## Section 4 — Split into Training and Test Sets
This is one of the standard patterns in machine learning.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print(X_train.shape)
print(X_test.shape)
```

## Section 5 — Train a Simple Classifier
We will use logistic regression, which is a standard baseline classifier.

```python
model = LogisticRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print(predictions)
```

## Section 6 — Evaluate the Model
We will look at accuracy and a classification report.

```python
acc = accuracy_score(y_test, predictions)
print("Accuracy:", acc)

print(classification_report(y_test, predictions, zero_division=0))
```

## Section 7 — Inspect Model Coefficients
This gives a simple view of which features are contributing to the prediction.

```python
coefficients = pd.DataFrame({
    "feature": X.columns,
    "coefficient": model.coef_[0]
})

coefficients
```

## Section 8 — Clustering
Unsupervised learning groups samples without using the responder label.

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X)

df["cluster"] = clusters
df
```

## Section 9 — PCA for Visualization
PCA reduces features into a small number of components that can be visualized.

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pca_df["sample"] = df["sample"]
pca_df["responder"] = df["responder"]
pca_df
```

### Exercises

## Section 10 — Plot the PCA Result
This is a simple embedding-style visualization.

```python
plt.figure(figsize=(6, 4))
plt.scatter(pca_df["PC1"], pca_df["PC2"])

for _, row in pca_df.iterrows():
    plt.text(row["PC1"], row["PC2"], row["sample"])

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of Toy Biological Features")
plt.show()
```
