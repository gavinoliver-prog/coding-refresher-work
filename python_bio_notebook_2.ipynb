## Section 3 — FASTA Format Basics

```python
fasta_text = ">seq1\nATGCGT\n>seq2\nTTTAAA\n"
with open("example.fasta", "w") as f:
    f.write(fasta_text)

sequences = {}

with open("example.fasta") as f:
    name = None
    seq = []
    for line in f:
        line = line.strip()
        if line.startswith(">"):
            if name:
                sequences[name] = "".join(seq)
            name = line[1:]
            seq = []
        else:
            seq.append(line)
    sequences[name] = "".join(seq)

print(sequences)
```

## Section 4 — Reverse Complement

```python
comp = {"A":"T","T":"A","G":"C","C":"G"}

def reverse_complement(seq):
    return "".join(comp[b] for b in reversed(seq))

print(reverse_complement("ATGCGT"))
```

## Section 5 — Sliding Windows (k-mers)

```python
def kmers(seq, k):
    for i in range(len(seq) - k + 1):
        yield seq[i:i+k]

list(kmers("ATGCGT", 3))
```

## Section 6 — k-mer Counting

```python
from collections import Counter

seq = "ATGCGTATG"
counts = Counter(kmers(seq,3))
print(counts)
```

## Section 7 — Motif Search

```python
def find_motif(seq, motif):
    positions = []
    for i in range(len(seq) - len(motif) + 1):
        if seq[i:i+len(motif)] == motif:
            positions.append(i)
    return positions

find_motif("ATGCGTATG", "ATG")
```

## Section 9 — Sequence Statistics

```python
def gc_content(seq):
    return sum(b in "GC" for b in seq) / len(seq)

gc_content("ATGCGT")
```
